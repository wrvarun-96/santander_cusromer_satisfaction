```python
import pandas as pd

cust_train=pd.read_csv(r'C:\Users\W R VARUN\Desktop\Data Science\MentorBoxCo\customer\train.csv')
cust_train
```

```python
X_test=pd.read_csv(r'C:\Users\W R VARUN\Desktop\Data Science\MentorBoxCo\customer\test.csv')
```

```python
cust_train=cust_train.drop('ID',axis=1)
X_test=X_test.drop('ID',axis=1)
```

```python
#Count of each class
import seaborn as sns
sns.countplot(cust_train['TARGET'])
```

```python
Y=cust_train['TARGET']
X=cust_train.drop('TARGET',axis=1)
```

```python
from sklearn.feature_selection import VarianceThreshold

#Have to drop columns which are having variance as 0 which will not add any importance to model 
var_thresh=VarianceThreshold(threshold=0.01)
var_thresh.fit(X)
#Columns which are not having varianceas 0
print(sum(var_thresh.get_support()))
```

```python
not_required_col = [col for col in X.columns if col not in X.columns[var_thresh.get_support()]]

print(len(not_required_col))
```

```python
#Dropping unwanted columns
X=X.drop(not_required_col,axis=1)
X_test=X_test.drop(not_required_col,axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,10))
correlation=X.corr()
sns.heatmap(correlation,annot=True)

def correlation(data,thres):
    corrs=set()
    corr_matrix=data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>thres:
                colname=corr_matrix.columns[i]
                corrs.add(colname)
    return corrs
corr_features=correlation(X,0.98)
len(corr_features)
```

```python
#Dropping highly correlated features
X=X.drop(corr_features,axis=1)
X_test=X_test.drop(corr_features,axis=1)
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_train=scaler.fit_transform(X)
scaled_test=scaler.transform(X_test)
X_train_new=pd.DataFrame(scaled_train, columns = X.columns)
X_test_new = pd.DataFrame(scaled_test, columns = X_test.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_train_new,Y,test_size=0.30,random_state=0)
```

### Logistic Regression Before Imbalance

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


logit = LogisticRegression().fit(x_train, y_train)
y_pred_train = logit.predict(x_train)
y_pred = logit.predict(x_test)
print('Logistic accuracy of train is {} and test is {}'.format((accuracy_score(y_train, y_pred_train)),(accuracy_score(y_test, y_pred))))
print(classification_report(y_train, y_pred_train))
print(classification_report(y_test, y_pred))
```

### Logistic Regression after treating imbalance

```python
##Data is Imbalanced so have to balnce the output label for better performance of model

from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_sample, y_sample = smote.fit_resample(X_train_new,Y)
logit = LogisticRegression().fit(x_sample, y_sample)
y_pred_train1= logit.predict(x_sample)
y_pred1 = logit.predict(x_test)
print('Logistic accuracy of train is {} and test is {}'.format((accuracy_score(y_sample, y_pred_train1)),(accuracy_score(y_test, y_pred1))))
print(classification_report(y_sample, y_pred_train1))
print(classification_report(y_test, y_pred1))
```

### USing PCA for reducing dimensonality

```python
# instantiate PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=90)
compo = pca.fit_transform(x_sample)
X_TEST=pca.transform(X_test_new)
```

```python
train_pcs = pd.DataFrame(data = compo)
Y_trains= pd.Series(y_sample, name='TARGET')
```

```python
print('\n Total Variance Explained:', round(sum(list(pca.explained_variance_ratio_))*100, 2))
```

```python
from sklearn.model_selection import train_test_split 
x_train_new,x_test_new,y_train_new,y_test_new=train_test_split(train_pcs,Y_trains,test_size=0.30,random_state=0)
```

### Logistic Reg after using PCA

```python
logs = LogisticRegression().fit(x_train_new, y_train_new)
y_pred_train=logs.predict(x_train_new)
y_pred_test = logs.predict(x_test_new)
print('Logistic accuracy of train is {} and test is {}'.format((accuracy_score(y_train_new, y_pred_train)),(accuracy_score(y_test_new, y_pred_test))))
print(classification_report(y_train_new, y_pred_train))
print(classification_report(y_test_new, y_pred_test))
```

### Xgboost AND Random_Forest

```python
from sklearn.ensemble import RandomForestClassifier
import xgboost

boost=xgboost.XGBClassifier().fit(x_train_new, y_train_new)
pred_train1=boost.predict(x_train_new)
pred_test1=boost.predict(x_test_new)
#print(boost.score(X_resampled,Y_resampled))
#print(boost.score(X_test,y_test))
print(classification_report(y_train_new,pred_train1))
print(classification_report(y_test_new,pred_test1))

forest=RandomForestClassifier().fit(x_train_new, y_train_new)
pred_train2=forest.predict(x_train_new)
pred_test2=forest.predict(x_test_new)
#print(forest.score(X_resampled,Y_resampled))
#print(forest.score(X_test,y_test))
print(classification_report(y_train_new,pred_train2))
print(classification_report(y_test_new,pred_test2))
```
